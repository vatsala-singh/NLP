```python
#scrape hindisamay website to get all the sublinks of article
def scrape(url):
    response = requests.get(url)
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.content, "html.parser")
        results = soup.find_all('span', {'id': 'ContentPlaceHolder1_lblRight'})
        print(results)
        for r in results:
            res = r.find_all('a')
        hyperlinks = [i['href'] for i in res]

    else:
        print("Failed to fetch the website")
    return hyperlinks
```

```python
url_harishankar_parsai = "https://www.hindisamay.com/writer/%E0%A4%B9%E0%A4%B0%E0%A4%BF%E0%A4%B6%E0%A4%82%E0%A4%95%E0%A4%B0-%E0%A4%AA%E0%A4%B0%E0%A4%B8%E0%A4%BE%E0%A4%88-.cspx?id=1174&name=%E0%A4%B9%E0%A4%B0%E0%A4%BF%E0%A4%B6%E0%A4%82%E0%A4%95%E0%A4%B0-%E0%A4%AA%E0%A4%B0%E0%A4%B8%E0%A4%BE%E0%A4%88-"
url_premchand = "https://www.hindisamay.com/writer/%E0%A4%AA%E0%A5%8D%E0%A4%B0%E0%A5%87%E0%A4%AE%E0%A4%9A%E0%A4%82%E0%A4%A6-.cspx?id=1152&name=%E0%A4%AA%E0%A5%8D%E0%A4%B0%E0%A5%87%E0%A4%AE%E0%A4%9A%E0%A4%82%E0%A4%A6-"
url_nirmal_verma = "https://www.hindisamay.com/writer/%E0%A4%A8%E0%A4%BF%E0%A4%B0%E0%A5%8D%E0%A4%AE%E0%A4%B2-%E0%A4%B5%E0%A4%B0%E0%A5%8D%E0%A4%AE%E0%A4%BE.cspx?id=1294&name=%E0%A4%A8%E0%A4%BF%E0%A4%B0%E0%A5%8D%E0%A4%AE%E0%A4%B2-%E0%A4%B5%E0%A4%B0%E0%A5%8D%E0%A4%AE%E0%A4%BE"
url_mahadevi_verma = "http://www.hindisamay.com/writer/%E0%A4%AE%E0%A4%B9%E0%A4%BE%E0%A4%A6%E0%A5%87%E0%A4%B5%E0%A5%80-%E0%A4%B5%E0%A4%B0%E0%A5%8D%E0%A4%AE%E0%A4%BE.cspx?id=1319&name=%E0%A4%AE%E0%A4%B9%E0%A4%BE%E0%A4%A6%E0%A5%87%E0%A4%B5%E0%A5%80-%E0%A4%B5%E0%A4%B0%E0%A5%8D%E0%A4%AE%E0%A4%BE"
```

```python
hyperlinks_harishankar_parsai = scrape(url_harishankar_parsai)
hyperlinks_premchand = scrape(url_premchand)
hyperlinks_nirmal_verma = scrape(url_nirmal_verma)
hyperlinks_mahadevi_verma = scrape(url_mahadevi_verma)
```

```python
# method to scrape the hyperlinks collected from scrape function
def content(hyperlinks):
    res_doc = []
    for h in hyperlinks:
        response = requests.get(h.strip())
        # Check if the request was successful
        if response.status_code == 200:
            # Parse the HTML content
            soup = BeautifulSoup(response.content, "html.parser")
            # Extract the desired content
            # Example: Extract all the text within <p> tags
            paragraphs = soup.find_all("p")
            result_string = ""
            for p in paragraphs:
                print(p.text)
                result_string = result_string + " " + p.text
            res_doc.append(result_string)
        else:
            print("Failed to fetch the website")
    return res_doc
```

```python
list_harishankar = content(hyperlinks_harishankar_parsai)
```
